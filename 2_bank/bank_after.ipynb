### Обработка пропусков в колонках `days_employed` и `total_income`

Как было выявлено ранее, в колонах `days_employed` (общий трудовой стаж в днях) и `total_income` (ежемесячный доход) заполнено только по 19351 строк из 21525. 

Так как кол-во уникальных значений в столбцах `days_employed` и `total_income` совпадает, а в других столбцах пропущенных значений нет, то выдвигаются предположения:
- каждое пропущенное значение в столбцах `days_employed` и `total_income` относится к одному и тому же уникальному заемщику (пропуски в обоих столбцах раcпределены по строкам одинаково/параллельно),
- значения в столбцах `days_employed` и `total_income` относятся к одной категории заемщиков.

Например, данные могли бы отсутствовать в отношении заемщиков, ранее никогда не работавших (студенты без опыта работы или проработавшие на первом месте менее 1-го месяца).

Сначала проверю распределение пропущенных строк по категориям заемщиков:

```python
data.pivot_table(index='income_type', values='total_income', aggfunc='count', margins=True).reset_index()
```

Количество уникальных значений в каждой категории заемщиков отличается от общего количества уникальных значений в колонке, а это значит, что пропущенные данные относятся к разным категориям заемщиков. Кроме того, стало очевидно, про пропущенных значения распределеных по всем категориям заемщиков, присутствующим в датасете.

Таким образом, получено подтверждение того, что пропуски в колонках `days_employed` и `total_income` не зависят от категории заемщика. Вероятнее всего пропуски действительно относятся к к одному и тому же уникальному заемщику, но так как в датасете отсутствует `user_id`, проверить это невозможно. Однако в рамках исследования я буду считать, что пропуски распределены по пользователям.

Теперь необходимо решить что делать с этими пропусками: удалять или заполнять. Для этого определю долю пропущенных значений:

```python
missed_str = data['days_employed'].isna().mean()

display(f'Доля пропущенных значений в "days_employed" составила {missed_str:.0%}')
```

Пропуски составляют 10% от всего датасета и это существенная величина, просто удалить данные будет грубой ошибкой.

Поэтому делается предположение о том, что стаж указан в часах (300000/24 = 12 500 дней/365 = 34 года). 
Напишу функцию для корректировки данных о трудовом стаже для выбранных категорий заемщиков:

Поэтому для устранения возрастной аномалии для всех строк, где возраст в колонке `check_years` меньше 18, он будет увеличен на 18 лет (совершенноление) от текущего значения общего трудового стажа в годах (new_days_employed/365).

Напишу функцию для изменения возраста:

```python
def age(row):
    check_years = row['check_years']
    dob_years = row['dob_years']
    if check_years <18:
        return int(row['new_days_employed']/365 + 18)
    else:
        return dob_years
```

Применю функцию и скорректирую порядок столбцов:

```python
data['new_dob_years'] = data.apply(age, axis=1)

data = data.loc[:, ['children', 'days_employed', 'new_days_employed', 'check_years', 'new_dob_years', 
                    'dob_years', 'education', 'education_id', 'family_status', 'family_status_id', 'gender',
                    'income_type', 'debt', 'total_income', 'purpose', 'purpose_category']]
data.head()
```

```python
data['new_check_years'] = data['new_dob_years']-(data['new_days_employed']/365) 

data['new_check_years'] = data['new_check_years'].astype('int')

data.head()
```

### Проверяем наличие зависимости между уровнем дохода и возвратом кредита в срок

Прежде чем приступать к поискам закономерностей, необходимо категоризировать доход заемщиков, чтобы с данными было удобно работать. 

Категории дохода и их стоимостной диапазан заданы условиями к исследованию:
- доход 0–30000 рублей —  категория'E';
- доход 30001–50000 рубле — категория 'D';
- доход 50001–200000 — категория 'C';
- доход 200001–1000000 — категория 'B';
- дозод 1000001 и выше — категория 'A'.

Создам функции для категоризации дохода заемщиков:

```python
# функция для присвоения категории дохода
def income_category (row):
    total_income = row['total_income']
    
    if 0 <= total_income <= 30000:
        return 'E'
    if 30001 <= total_income <= 50000:
        return 'D'
    if 50001 <= total_income <= 200000:
        return 'C'
    if 200001 <= total_income <= 1000000:
        return 'B'
    if total_income >= 1000001:
        return 'A'
    
# функцию с расшифровкой стоимостного диапазона категорий дохода

def income (row):
    total_income = row['total_income_category']
    
    if total_income == 'E':
        return 'менее 30 000'
    if total_income == 'D':
        return 'от 30 000 до 50 000'
    if total_income == 'C':
        return 'от 50 000 до 200 000'
    if total_income == 'B':
        return 'от 200 000 до 1 000 000'
    if total_income == 'A':
        return 'более 1 000 000'
```

Применю функции:

```python
# применение функции для категоризации дохода заемщиков

data['total_income_category'] = data.apply(income_category, axis=1)

# применение функции для расшифровки стоимостного диапазона категорий дохода
    
data['income'] = data.apply(income, axis=1)

data.head()
```
