### Приоритизация гипотез

Метод приоритизации ICE рассчитывается как произведение Impact, Confidence и Ease или как частное от произведения Impact и Confidence, разделенного на Efforts.
Рассчитаем этот показатель для каждой гипотезы, исходя из имеющихся показателей.

```python
data['ice'] = (data['impact'] * data['confidence'])/ data['efforts']
```

Метод приоритизации RICE рассчитывается как частное от произведения Reach, Impact и Confidence, разделенного на Efforts. Рассчитаем этот показатель для каждой гипотезы, исходя из имеющихся показателей.

```python
data['rice'] = (data['reach'] * data['impact'] * data['confidence'])/data['efforts']
```

Сравним результаты методов ICE и RICE

```python
data[['hypothesis','ice','rice']].style.background_gradient(low=0.1, high=0, axis=0)
```

Сравним списки потенциально наиболее успешных гипотез:
 - ICE - 8, 0, 7 и 6.
 - RICE - 7, 2, 0 и 6.

Как мы видим, в обоих случаях в топ-3 есть совпадения в гипотезах под номерами 0 и 7, а также неплохие показатели у гипотезы 6, но её стоит рассматривать скорее во вторую очередь.

При использовании метода RICE разница возникает в отношении гипотезы 8, которая сформулирована так: "Запустить акцию, дающую скидку на товар в день рождения". Несмотря на то, что в обоих случаях оценка гипотезы имеет одинаковое значение, для метода RICE она находится на 5-м месте. Это объясняется тем, метод RICE, помимо прочего, учитывает влияние параметра Reach, отвечающего за количество пользователей, которых затронет внедряемое изменение.

Так как гипотезы приоретизируются для будущего А/В-тестирования, которое ограничено по времени, то в таких условиях данные изменения охватят значительно меньшее число пользователей, поэтому у гипотезы 8 фактор Reach имеет минимальную оценку влияния. Следовательно, при выборе этой гипотезы выборка может получиться недостаточно релевантной и результаты будут искажены.

Нельзя сказать, что один метод хуже другого, но объективно метод RICE можно считать более надежным, так как он учитывает риски тестирования гипотез с бОльшим количеством факторов.

По результатам двух методов наиболее перспективными являются гипотезы под номерами 7, 0 и 6 (гиптезы указаны в порядке убывания). Далее в рамках проекта будет рассматриваться гипотеза 7.

#### Создание датафрейма с кумулятивными данными

Для построения различных графиков кумулятивных показателей необходимо создать датасет с соответствующими данными, поэтому сначала подготовим данные для него. Во-первых, необходимо создать массив уникальных пар значений дат и групп теста

```python
dates_groups = orders[['date','group']].drop_duplicates()

dates_groups.head()
```

Затем нужно собрать агрегированные кумулятивные по дням данные о заказах.

```python
orders_agg = (dates_groups
              .apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]
              .agg({'date' : 'max', 'group' : 'max', 'transaction_id' : 'nunique', 'visitor_id' : 'nunique', 
                    'revenue' : 'sum'}), axis=1)
              .sort_values(by=['date','group']))

orders_agg.head()
```

Далее аналогично получим агрегированные кумулятивные по дням данные о посетителях интернет-магазина.

```python
visitors_agg = (dates_groups
                .apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], 
                                                         visitors['group'] == x['group'])]
                .agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1)
                .sort_values(by=['date','group']))

visitors_agg.head()
```

Теперь создадим датасет с кумулятивными данными: объединим кумулятивные данные о заказах и визитах в одной таблице и переименуем столбцы

```python
cumul_data = orders_agg.merge(visitors_agg, left_on=['date', 'group'], right_on=['date', 'group'])

cumul_data.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

cumul_data.head()
```

Датафрейм готов, осталось только проверить совпадают ли мимальная и максимальная даты в кумулятивном датафрейме с мимальной и максимальной датами в исходных датафреймах. Для начала определим минимальные и максимальные даты.

```python
# определим минимальную и максимальную даты оформления заказов

min_orders_date = orders['date'].min()
max_orders_date = orders['date'].max()

display('Самая ранняя дата оформления заказа {}'.format(min_orders_date))
display('Самая поздняя дата оформления заказа {}'.format(max_orders_date))
```

```python
# определим минимальную и максимальную даты посещения сайта

min_visitors_date = visitors['date'].min()
max_visitors_date = visitors['date'].max()

display('Самая ранняя дата посещения сайта {}'.format(min_visitors_date))
display('Самая поздняя дата посещения сайта {}'.format(max_visitors_date))
```

```python
# определим минимальную и максимальную даты кумулятивных данных

min_cumul_date = cumul_data['date'].min()
max_cumul_date = cumul_data['date'].max()

display('Самая ранняя дата {}'.format(min_cumul_date))
display('Самая поздняя дата {}'.format(max_cumul_date))
```

Даты определены, теперь мы можем сравнить их.

```python
# сравним минимальную дату

if min_orders_date == min_visitors_date == min_cumul_date:
    display('Начальная дата сбора данных совпадает во всех датафреймам и равна {}'.format(min_orders_date))
else:
    display('Начальные даты сбора данных не равны')
```

```python
# сравним максимальную дату

if max_orders_date == max_visitors_date == max_cumul_date:
    display('Дата окончания сбора данных совпадает во всех датафреймам и равна {}'.format(max_orders_date))
else:
    display('Даты окончания сбора данных не равны')
```

#### Расчет статистической значимости различий в конверсии между группами по «сырым» данным

Для того, чтобы оценить, если ли разница в поведении пользователей тестируемых групп,а следовательно, есть ли эффект от внедрения тестируемой гипотезы, необходимо сравнить выборки между собой.

Первое, что необходимо для этого сделать - сформулировать гипотезы сравнения выборк, второе - подготовить данные и последнее - сравненить гипотез.

Итак, сформулируем гипотезы сравнения выборок:

 - нулевая: различий в среднем количестве заказов (конверсии) между группами нет, 
 - альтернативная: различия в конверсии между группами есть.

Далее приступим к подготовке данных. Нам понодобится общий датафрейм с ежедневными и кумулятивными по дням сведениями по количестве заказов, их стоимости и количестве визитов по каждой группе тестирования. Кумулятивные данные у нас  уже хранятся в переменных cumul_data_a и cumul_data_b, поэтому начнем со сбора данных по дням.

Первым шагом определим количество посещений сайта по дням отдельно по группам:

```python
# определим количество посещений в группе А
visitors_a_daily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitors_a_daily.columns = ['date', 'visitors_date_a']

visitors_a_daily.head()
```

```python
# определим количество посещений в группе B
visitors_b_daily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitors_b_daily.columns = ['date', 'visitors_date_b']

visitors_b_daily.head()
```

Вторым шагом посчитаем количество заказов и их стоимость также по дням для каждой группы:

```python
# определим количество заказов и их стоимость в группе А
orders_a_daily = (orders[orders['group'] == 'A'][['date', 'transaction_id', 'visitor_id', 'revenue']]
                  .groupby('date', as_index=False)
                  .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'}))

orders_a_daily.columns = ['date', 'orders_date_a', 'revenue_date_a']

orders_a_daily.head()
```

```python
# определим количество заказов и их стоимость в группе И
orders_b_daily = (orders[orders['group'] == 'B'][['date', 'transaction_id', 'visitor_id', 'revenue']]
                  .groupby('date', as_index=False)
                  .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'}))

orders_b_daily.columns = ['date', 'orders_date_b', 'revenue_date_b']

orders_b_daily.head()
```

Теперь скорректируем кумулятивные данные, так как в созданных датафреймах cumul_data_a и cumul_data_b содержится больше информации, чем понадобится нам на данном этапе.

```python
# выделим из cumul_data_a необходимые данные по группе А
cumul_a = cumul_data_a.loc[:,['date', 'orders', 'revenue', 'visitors']]

cumul_a.columns = ['date', 'cumul_orders_a', 'cumul_revenue_a', 'cumul_visitors_a']

cumul_a.head()
```

```python
# выделим из cumul_data_b необходимые данные по группе B
cumul_b = cumul_data_b.loc[:,['date', 'orders', 'revenue', 'visitors']]

cumul_b.columns = ['date', 'cumul_orders_b', 'cumul_revenue_b', 'cumul_visitors_b']

cumul_b.head()
```

Все исходные данные подготовлены и мы можем создать общий датафрейм, объединив все датасеты:

```python
o_v_cumul_daily = (orders_a_daily.merge(visitors_a_daily, left_on='date', right_on='date', how='left')
                   .merge(orders_b_daily, left_on='date', right_on='date', how='left')
                   .merge(visitors_b_daily, left_on='date', right_on='date', how='left')
                   .merge(cumul_a, left_on='date', right_on='date', how='left')
                   .merge(cumul_b, left_on='date', right_on='date', how='left'))

o_v_cumul_daily.head(5)
```

Далее для расчетов необходимо в каждой группе отобрать уникальных пользователей и заполнить по ним информацию о факте совершения или не совершения заказов по принципц булева - для тех, кто оформил хотя бы один заказ будет указана 1, а для тех, кто не оформил ни одного заказа - 0.

Для формирования таких данных мы не можем использовать уже существующую переменную orders_by_users, так как в ней нет разделения на группы тестирования. Именно по этому необходимо создать новые переменные, которые будут соответствовать заданным условиям:

```python
# определим число заказов для пользователей группы А, которые совершили хотя бы 1 заказ
orders_by_users_a = (orders[orders['group'] == 'A']
                    .groupby('visitor_id', as_index=False)
                    .agg({'transaction_id': pd.Series.nunique}))

orders_by_users_a.columns = ['visitor_id', 'orders']

orders_by_users_a.head()
```

```python
# определим число заказов для пользователей группы В, которые совершили хотя бы 1 заказ
orders_by_users_b = (orders[orders['group'] == 'B']
                    .groupby('visitor_id', as_index=False)
                    .agg({'transaction_id': pd.Series.nunique}))

orders_by_users_b.columns = ['visitor_id', 'orders']

orders_by_users_b.head()
```

Затем для пользователей каждой группы необходимо составить список без указания id пользователя, где каждый элемент — число заказов определённого пользователя, в том числе ноль. Это нужно, чтобы подготовить выборки к проверке критерием Манна-Уитни.

Чтобы составить такой списк, необходимо создать объект pd.Series и сохранить в нем выборку с количеством заказов каждого пользователя, для чего у нас уже сохранены данные в переменных orders_by_users_a и orders_by_users_b.

Осталось решить, как определить пользователей, у которых должно быть указано нулевое количество заказов?
Количество таких пользователей равно разнице между суммой всех посетителей за день и количеством записей о заказах в этот день. Вычислим эти данные в формате pd.Series и объединим два списка.

Таким образом, число элементов созданного списка в формате pd.Series будет равно количеству пользователей, а сумма всех элементов — количеству заказов.

```python
# вычислим количество заказов для всех пользователей группы А (сырые данные)
sample_a = pd.concat([orders_by_users_a['orders'],
           pd.Series(0, index=np.arange(o_v_cumul_daily['visitors_date_a'].sum() - len(orders_by_users_a['orders'])),
           name='orders')], axis=0)
```

```python
# вычислим количество заказов для всех пользователей группы В (сырые данные)
sample_b = pd.concat([orders_by_users_b['orders'],
           pd.Series(0, index =np.arange(o_v_cumul_daily['visitors_date_b'].sum() - len(orders_by_users_b['orders'])),
           name='orders')], axis=0)
```

Теперь мы можем приступить к расчету статистической значимости различий в конверсии по сырым данным (без удаления аномальных значений).

В случаях, когда в данных есть большие (по сравнению с нормальным распределнием) выбросы, для оценки статистической значимости различий применяют непараметрическте методы, то есть методы, основанные на структурном подходе, а не на алгебраических метриках (так как при значительных аномалиях последние существенно искажены).

Оценим выборку заказов методом Уилкоксона-Манна-Уитни по среднему числу заказов на пользователя.

```python
alpha = 0.05  # критический уровень статистической значимости

results_conv = stats.mannwhitneyu(sample_a, sample_b)

display('p-значение: {0:.5f}'.format(results_conv.pvalue))

if results_conv.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    display('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

Вспомним выдвинутые гипотезы:
 - нулевая: различий в среднем количестве заказов (конверсии) между группами нет,
 - альтернативная: различия в конверсии между группами есть.
 
Полученное значение p-valu, равное 1,6% меньше критического уровня значимости, равного 5%. Таким образом, анализ выборки свидетельствует нам о разном поведении пользователей групп А и В в период тестирования, а значит, ранее выдвинутые предположения о том, что показатели группы В превосходят показатели группы А подтверждается. Однако необходимо понимать, что данные расчеты проведены по сырым данным и для полноты сведений требуется проверка по очищенным данным. 

Вычислим относительный прирост среднего числа заказов группы B по отношению к группе А за период тестирования:

```python
growth_conv = (sample_b.mean() / sample_a.mean() - 1)

"Относительный прирост среднего числа заказов группы В по отношению к группе А составил {0:.2%}".format(growth_conv)
```

Сравнение по сырым данным показывает значительное опережение темпов роста конверсии группы В по сравнению с группой А, что коррелирует с тенденцией на графике относительного изменения конверсии группы В к группе А, однако, необходимо сравнить эти результаты с расчетами по очищенным данным для принятия решения о результатах тестирования.

#### Расчет статистической значимости различий в среднем чеке заказа между группами по «сырым» данным

Аналогичным образом требуется оценить метрику среднего чека - необходимо проверить, существует ли статистически значимая разница между группами А и В.
Сформулируем гипотезы:
 - нулевая гипотеза: различий в среднем чеке между группами нет,
 - альтернативная гипотеза: различия в среднем чеке между группами есть.

Чтобы рассчитать статистическую значимость различий в среднем чеке, рассмотрим данные о выручке с заказов:

```python
alpha = 0.05  # критический уровень статистической значимости

results_rev = stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'])

display('p-значение: {0:.5f}'.format(results_rev.pvalue))

if results_rev.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    display('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

Полученный результат p-value в 72,9% существенно превышает установленный критический уровень статистической значимости. Это значит, что мы не обнаружили статистически значимых различий в выборках группы А и В при сравнении показателей среднего чека, поэтому у нас нет причин отвергать нулевую гипотезу. 

Пока по сырым данным нам следует придерживаться утверждения, что различий в среднем чеке между группами нет: то есть, изменения на сайте не привели к изменению модели поведения пользователей при совершении покупок и увеличению среднего чека.

Посмотрим на относительный прирост среднего чека группы B по отношению к группе А за период тестирования:

```python
growth_rev = (orders[orders['group']=='B']['revenue'].mean()/orders[orders['group']=='A']['revenue'].mean())-1

"Относительный прирост среднего чека группы В по отношению к группе А составил {0:.2%}".format(growth_rev)
```

В относительных величинах показатель среднего чека группы B оказался на четверть выше среднего чека группы A, что коррелирует с графиков относительного изменения среднего чека группы В к А. Но необходимо помнить, что график кумулятивного среднего чека позволил выдвинуть предположении о наличии в данных аномальных пользователей, существенно влияющих на результаты тестирования. Поэтому необходимо проверить, как изменится показатель среднего чека при оценке по очищенным данным.

### Анализ А/В-теста по очищенным данным

#### Выделение аномальных пользователей

Для расчетов по очищенными данным необходимо сначала их подготовить, а для этого требуется вспомнить, какие критерии были ранее определены в исследовании для отсечения аномалий:
 - пользователи со стоимостью заказов свыше 28000 (95-й перцентиль),
 - пользователи, совершившие 3 заказа и более (95-й перцентиль).
 
Это позволит удалить по 5% пользователей с наибольшим числом заказов и пользователей с дорогими заказами, что укладывается в общепринятую практику отсечения от 1% до 5% аномальных данных. Подготовим срезы пользователей по заданным параметрам:

```python
# срез аномальных пользователей по числу заказов
users_with_many_orders = pd.concat([orders_by_users_a[orders_by_users_a['orders'] > p95_orders]['visitor_id'],
                                    orders_by_users_b[orders_by_users_b['orders'] > p95_orders]['visitor_id']], axis=0)

# срез аномальных пользователей по стоимости заказов
users_with_expensive_orders = orders[orders['revenue'] > p95_rev]['visitor_id']

abnormal_users = (pd.concat([users_with_many_orders, users_with_expensive_orders], axis=0)
                 .drop_duplicates().sort_values())

display(abnormal_users.head(5))
display(abnormal_users.shape[0]) 
```

#### Создание датафрейма с очищенными кумулятивными данными

```python
orders_filtered = orders[np.logical_not(orders['visitor_id'].isin(abnormal_users))]
```

Итак, исходный датафрейм orders очищен от аномальных пользователей, теперь создадим датасет с кумулятивными данными:

```python
orders_agg_filtered = (dates_groups
                      .apply(lambda x: orders_filtered[np.logical_and (orders_filtered['date'] <= x['date'], 
                                                                       orders_filtered['group'] == x['group'])]
                      .agg({'date' : 'max', 'group' : 'max', 'transaction_id' : 'nunique', 
                            'visitor_id' : 'nunique', 'revenue' : 'sum'}), axis=1).sort_values(by=['date','group']))

orders_agg_filtered.columns = ['date', 'group', 'orders', 'buyers', 'revenue']

orders_agg_filtered.head()
```

#### Анализ графика кумулятивной выручки по группам (по очищенным данным)

Для сравнения посмотрим как изменился график кумулятивной выручки, если построить его по очищенным данным. Для построения графика выделим показатели групп в отдельные переменные:

#### Расчет статистической значимости различий в конверсии между группами по «очищенным» данным

Мы определили количество аномальных пользователей и их уникальные номера. И теперь мы сможем оценить их влияние на результаты тестирования, оценив статистическую значимость различий в метриках групп по очищенным данным. Как всегда начнем с подготовки выборки:

```python
# вычислим количество заказов для всех пользователей группы А (очищенные данные)
sample_a_filtered = pd.concat([orders_by_users_a[np.logical_not(orders_by_users_a['visitor_id']
        .isin(abnormal_users))]['orders'],
        pd.Series(0,index=np.arange(o_v_cumul_daily['visitors_date_a'].sum() - len(orders_by_users_a['orders'])),
        name='orders')], axis=0)

# вычислим количество заказов для всех пользователей группы В (очищенные данные)
sample_b_filtered = pd.concat([orders_by_users_b[np.logical_not(orders_by_users_b['visitor_id']
        .isin(abnormal_users))]['orders'],
        pd.Series(0,index=np.arange(o_v_cumul_daily['visitors_date_b'].sum() - len(orders_by_users_b['orders'])),
        name='orders')], axis=0) 
```

Применим статистический критерий Манна-Уитни к полученным выборкам:

```python
alpha = 0.05  # критический уровень статистической значимости

results_conv_filtered = stats.mannwhitneyu(sample_a_filtered, sample_b_filtered)

display('p-значение: {0:.5f}'.format(results_conv_filtered.pvalue))

if results_conv_filtered.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    display('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

Значение p-value изменилось незначительно: оно снизилось с 1,6% до 1,3%, то есть, оно по-прежнему меньше критического уровня значимости в 5% и нулевую гипотезу необходимо отвергнуть.

Вспомним гипотезы:
 - нулевая: различий в среднем количестве заказов (конверсии) между группами нет, 
 - альтернативная: различия в конверсии между группами есть.

Следовательно, в выборке продолжает сохраняться статистически значимая разница, подтверждающая отличное поведение пользователей в двух группах.

```python
growth_conv_filtered = (sample_b_filtered.mean() / sample_a_filtered.mean() - 1)

"Относительный прирост среднего числа заказов группы В по отношению к группе А составил {0:.2%}".format(growth_conv_filtered)
```

Относительный показатель динамики конверсии группы В по отношению к группе А увеличился с 13,81% до 17,27%. Это означает, что конверсия в группе В растет быстрее, чем в группе А на 17,27%, что подтверждает альтернативную теорию о наличии различий в конверсии между группами.

#### Расчет статистической значимости различий в среднем чеке заказа между группами по «очищенным» данным

Проверим, изменились ли значения p-value и показатель прироста в отношении метрики среднего чека заказа.

```python
alpha = 0.05  # критический уровень статистической значимости

results_rev_filtered = stats.mannwhitneyu(orders[np.logical_and(orders['group'] == 'A',
                np.logical_not(orders['visitor_id'].isin(abnormal_users)))]['revenue'], 
                orders[np.logical_and(orders['group'] == 'B', 
                np.logical_not(orders['visitor_id'].isin(abnormal_users)))]['revenue'])

display('p-значение: {0:.5f}'.format(results_rev_filtered.pvalue))

if results_rev_filtered.pvalue < alpha:
    display('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    display('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя')
```

Показатель p-value составил 73,8%, что всего лишь на 0,9% больше, чем при оценке по сырым данным, Мы снова получили подтверждение о том, что статистически значимой разницы между группой А и группой В в отношении метрики среднего чека нет.

Посмотри, как изменился относительный показатеть метрики - прирост среднего чека группы В по отношению к группе А:

```python
growth_rev = ((orders[np.logical_and(orders['group'] == 'B',
                np.logical_not(orders['visitor_id'].isin(abnormal_users)))]['revenue'].mean())
                / (orders[np.logical_and(orders['group'] == 'A',
                np.logical_not(orders['visitor_id'].isin(abnormal_users)))]['revenue'].mean())) - 1

"Относительный прирост среднего чека группы В по отношению к группе А составил {0:.2%}".format(growth_rev)
```

На очищенных данных становится очевидным, что темпы роста среднего чека группы В отстают от темповы роста группы А на 1,96%, что укладывается в проведенную выше проверку на выявление статистически значимых различий. В рамках А/В-тестирования этот показатель следует воспринимать как крайне незначительное изменение поведения пользователей, то есть в отношении данной метрики гипотеза не работает (изменения можно считать органическими) либо работает крайне слабо.

Такие изменения в объемах выручки могут быть как за счет увеличения числа заказов (повышения конверсии посетителей сайта в заказчиков и/или повышения конверсии повторных заказов), так и за счет увеличения доли очень дорогих заказов в выборке (аномалий).

Далее, при помощи диаграмм рассеивания (`scatter()`) и расчета перцентилей (`percentile()`) были определены пороговые значения аномальных пользователей:

И последней частью расчетов данного раздела является определение статистически значимой разницы (`mannwhitneyu()`) в конверсии и в среднем чеке между группами:

Таким образом, анализ выборки свидетельствует нам о разном поведении пользователей групп А и В в период тестирования, а значит, ранее выдвинутые предположения о том, что показатели группы В превосходят показатели группы А подтверждается. В относительных величинах конверсия в группе В растет быстрее, чем в группе А на 13,81%. 

2. В отношение среднего чека сравнение выборки по группам подтвердило отсутствие статистически значимого различия. Полученный результат p-value в 72,9% существенно превышает установленный критический уровень статистической значимости (5%). Это значит, что мы не обнаружили статистически значимых различий в выборках группы А и В при сравнении показателей среднего чека, поэтому у нас нет причин отвергать нулевую гипотезу об отсутствии различий в среднем чеке между группами. Тем не менее в относительных величинах показатель среднего чека группы B оказался на четверть выше среднего чека группы A.
