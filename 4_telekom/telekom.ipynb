# Проект по определению выгодного тарифа для телеком компании

**Цель исследования** - определить, какой из двух тарифов, `Smart` или `Ultra`, приносит больше выручки федеральному оператору сотовой связи.

Параметры тарифных планов:
- smart - 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика, абонентская плата 550 руб.
- ultra - 3000 минут, 1000 сообшений и 30 Гб, абонентская плата 1950 руб.

**Задачи исследования:**
1. Сравнить ежемесячное потребление услуг для каждого тарифа.
2. Проверить гипотезы о средней выручке:
    - средняя выручка пользователей тарифов `Ultra` и `Smart` различается.
    - средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

**Источник данных** - 5 датасетов с данными о 500-х пользователях оператора за 2018 год. Датасеты содержат информацию о пользователях, о потребленных ими минутах, сообщениях, трафике и об их интернет-сессиях.

**Навыки и инструменты** - python, pandas, matplotlib, numpy, scipy, описательная статистика, проверка статистических гипотез.
____

# Резюме

Предположительно тарифные планы предназначены для разных целевых аудиторий клиентов - `Smart` более доступный и массовый продукт, а `Ultra` рассчитан на клиентов с более специфическими или узкоспециализированными требованиями к связи.

Тариф `Ultra` стоит в 3,5 раза дороже, чем тариф `Smart` и, несмотря на то, что темпы прироста клиентов по тарифу `Smart` на 50% и более превышают темпы прироста по тарифу `Ultra`, последний ежемесячно приносит на 26%-63% выручки больше. 

Средний объем выручки в Москве практически не отличается от других регионов: только в 4-х месяцах выручках по Москве была несколько выше, чем в других регионах, но всего на 1-7%.

Таким образом, при сохранении выявленных темпов прироста новых клиентов, тариф `Ultra` более выгоден для оператора связи.

### Суммирование объемов потребленния услуг по месяцам

#### Суммирование данных о звонках

Цель исследования - определить лучший тариф, то есть тот, который приносит больше выручки компании.

Поэтому ниже данный раздел будет посвещен расчетам объемов потребленных услуг и выручки.

```python
# проверка пустых значений

good_data['gb_used'].isna().sum()
```

Теперь можно считать ежемесячную выручку по каждому клиенту. Напишем функцию по строке, которая будет проводить нужные вычисления:

```python
# напишем функцию для расчета выручки

def revenue(row):
    rub_monthly_fee = row['rub_monthly_fee']
    
    duration = row['month_duration']
    minutes_included = row['minutes_included']
    rub_per_minute = row['rub_per_minute']

    
    message_used = row['count_message']
    messages_included = row['messages_included']
    rub_per_message = row['rub_per_message']

    
    one_gd = 1024
    gb_used = row['gb_used']
    gb_included = row['mb_per_month_included']/one_gd
    rub_per_gb = row['rub_per_gb']

    result = rub_monthly_fee
    
    if duration > minutes_included:
        result += (duration - minutes_included)*rub_per_minute
        
    if message_used > messages_included:
        result += (message_used - messages_included)*rub_per_message
        
    if gb_used > gb_included:
        result += (gb_used - gb_included)*rub_per_gb
        
    return result

good_data['revenue'] = good_data.apply(revenue, axis=1)

good_data.head(10)
```

Разница между средним/медианным и максимальным значением более чем в 4 раза, поэтому нужно проверить полученный результат на аномалии (артефакты).

## Проверка гипотез

### Проверка нормальности распределения выборки

Прежде чем приступать к проверке гипотез важно проверить выборку на нармальность распределения данных, ведь в случае не нормального распределения проверка гипотез будет невозможна.

Для проверки нормальности распределения выборки помимо построения графиков необходимо рассчитать среднее, дисперсию и стандартное отклонение для каждого вида услуг по каждому тарифу, то есть 6 раз произвести одни и те же вычисления. 

Удобнее всего будет сначала выполнить расчеты, а потом проанализировать графики по каждой услуге. Для рассчета среднего, дисперсии и отклонения используем цикл и оформим результаты в таблицу:

```python
table = None

for t in ['month_duration', 'count_message', 'gb_used']:
    mean = good_data.groupby('tariff')[t].agg(['mean'])
    std = good_data.groupby('tariff')[t].agg(['std'])
    var = good_data.groupby('tariff')[t].agg(['var'])

    metric = (mean.merge(std, on='tariff').merge(var, on='tariff')).reset_index()
    metric['type'] = t
    metric = metric.reindex(columns=['type','tariff','mean','std','var'])

    if table is None:
        table = metric
    else:
        table = table.append(metric)

table = table.reset_index(drop=True)

table
```

### Проверка гипотезы о средней выручке пользователей тарифов «Ультра» и «Смарт»

Сформулируем гипотезы:
1. Нулевая гипотеза - средняя выручка пользователей тарифов «Ультра» и «Смарт» одинаковая (равна).
2. Альтернативная гипотеза - средняя выручка пользователей тарифов «Ультра» и «Смарт» различается.

При формулировании нулевой гипотезы опираемся на правило, согласно которому нулевая гипотеза должна быть сформулирована так, чтобы использовать знак равенства. В исходном задании к проекту нам была дана гипотеза "средняя выручка пользователей тарифов «Ультра» и «Смарт» различается". В соответствии с вышеуказанным правилом было решено считать данную теорию альтернативной. Исходя из этого и правила о равенстве, была сформулирована нулевая гипотеза.

Для получения расчетной величины вероятности получения наблюдаемого результата при условии, что нулевая гипотеза верна, определим пороговое значение alpha равным 5%, так как оно больше, а значит, вероятность ошибки в расчете по выборке меньше.

```python
# подготовим Series по тарифам для сравнения

good_data_smart_revenue = good_data.query('tariff == "smart"')['revenue']
good_data_smart_revenue.head()
```

```python
# подготовим Series по тарифам для сравнения

good_data_smart_revenue = good_data.query('tariff == "smart"')['revenue']

good_data_ultra_revenue = good_data.query('tariff == "ultra"')['revenue']

good_data_ultra_revenue.head()
```

```python
sample_1 = good_data_smart_revenue
sample_2 = good_data_ultra_revenue

alpha = 0.01
results = st.ttest_ind(sample_1, sample_2, equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Мы получили значение p-value, равное 4.2606313931076085e-250, значит, вероятность верности нашей нулевой гипотеза - 0%. Значит, мы можем быть уверены в том, что средняя выручка пользователей тарифов «Ультра» и «Смарт» точно не одинаковая.

При формулировании нулевой гипотезы опираемся на правило, согласно которому нулевая гипотеза должна быть сформулирована так, чтобы использовать знак равенства. В исходном задании к проекту нам была дана гипотеза "средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов". В соответствии с вышеуказанным правилом было решено считать данную теорию альтернативной. Исходя из этого и правила о равенстве, была сформулирована нулевая гипотеза.

Для получения расчетной величины вероятности получения наблюдаемого результата при условии, что нулевая гипотеза верна, определим пороговое значение alpha равным 5%, так как оно больше,  так как оно больше, а значит вероятность ошибки в расчете по выборке меньше.

Создадим таблицу для расчета выручки по регионам:

```python
good_data_city = good_data.copy()

good_data_city.head()
```

Так как в гипотезе у нас всего 2 города (Москва и все остальные), то заменим все отличные от Москвы города на термин "Другой":

```python
# напишем функцию для замены значений городов

def new_city(row):
    city = row['city']
    if 'Москва' in city:
        return city
    else:
        return 'Другой'

good_data_city['new_city'] = good_data_city.apply(new_city, axis=1)

good_data_city.head()
```

```python
good_data_other_revenue = good_data_city.query('new_city == "Другой"')['revenue']

good_data_other_revenue.head()
```

```python
sample_1 = good_data_moscow_revenue
sample_2 = good_data_other_revenue

alpha = 0.05
results = st.ttest_ind(sample_1, sample_2, equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Мы получили значение p-value, равное 0.5257376663729298, значит, с вероятностью в 52% наша нулевая гипотеза верна и выручка пользователей из Москвы не отличается от выручки пользователей регионов.
