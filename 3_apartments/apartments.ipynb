# Проект по анализу рынка недвижимости в Санкт-Петербурге и ЛО

**Цель исследования** - выявить типичные параметры объектов недвижимости, определяющие их рыночную стоимость.

**Задачи исследования:**
1. выполнить предобработку данных,
2. найти особенности и зависимости, характерные для рынка недвижимости.

**Источник данных** - один датасет с архивом объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах по данным ресурса Яндекс.Недвижимость за период с 2014 по 2019 года.

**Навыки и инструменты** - python, pandas, matplotlib, math, исследовательский анализ данных, визуализация данных, предобработка данных.
____

# Резюме

В общем случае, типичными и наиболее значимыми параметрами объектов недвижимости при определении их рыночной стоимоcти, являются (в порядке убывания влияния):
1. Площадь объекта.
2. Стоимость за квадратный метр. 
3. Комнатность объекта.
4. Удаленность от центра. 

Для объектов, расположенных в центре Санкт-Петербурга, факторы 3 и 4 меняются местами в степени оказываемого влияния на стоимость квартиры.

Кроме того, исследование позволило выявить сложную, нелинейную зависимость между стоимостью квартиры и такими параметрами как этажность и/или наличие поблизости парков/водоемов.

Влияние же временнОго фактора (дня недели, месяца или года публикации) в рамках исследования признано отсутствующим, так как по имеющимся данным зависимость удалось установить только с месяцем публикации, что скорее связано с некоторой сезонностью в жизни города.

По средним показателям объекты в центре Санкт-Петербурга отличаются от объектов иных локаций:
1. бОльшей площадью квартир - средний показатель 72 кв.м. на 25% больше по сравнению с другими районами города (58 кв.м.) и на 33-36% больше, чем в других локациях (53-54 кв.м.).
2. более высокой стоимостью квартир - средняя стоимость равна 8,9 млн, что на 2,6 млн или на 33% больше, чем в других районами города, и на 4,2 млн или на 52% больше по сравнению с городами и посёлками ЛО.
3. бОльшим количеством комнат - средний показатель равен 2,4 против 2,0 в других районах города и городах ЛО и против 1,7 в посёлках.
3. более высокими потолками - средний показатель равен 2.8м против 2.7м во всех других локациях.

# Краткое содержание работы

**_Первичное изучение датасета_** показало, что в исходных данных из 21 столбца полностью были заполнены только 8, а в 13-ти данные частично отсутствовали. Разброс отсутствующих данных широкий - от 49 до 15 620 строк. Наименьшее количество пропусков было допущено в столбцах `locality_name` и `floors_total` (менее 0,5% выборки). Далее из важных критериев оценки жилья не хватало примерно по 10% информации о жилой площади (`living_area`) и о площади кухни (`kitchen_area`). Наибольшее количество данных (из ключевых параметров) отсутствовало по высоте потолков (`ceiling_height`) - 38% датасета. 

На **_этапе предобработки_** были восстановлены следующие параметры:
1. локации объектов  в `locality_name`&
2. площади в `living_area` и `kitchen_area`&
3. высота потолков в `ceiling_height`

Соответственно, в случаях, когда для квартир со свободной планировкой не указана площадь кухни, было решено заполнить их нулевыми значениями. Через функцию заменим площадь кухни в студиях на 0.0:

```python
  def new_kitchen_studio(row):
    studio = row['studio']
    kitchen_area = row['kitchen_area']
    living_area = row['living_area']
    total_area = row['total_area']
    if studio == True and math.isnan(living_area):
        row['kitchen_area'] = 0.0
        row['living_area'] = total_area
    if studio == True:
        row['kitchen_area'] = 0.0
    else:
        kitchen_area
    return row

data = data.apply(new_kitchen_studio, axis=1)
```

Через функцию в квартирах свободной планировки заменим площадь кухни на 0.0, а площадь жилой приведем к общей площади:

```python
def new_open_plan(row):
    open_plan = row['open_plan']
    kitchen_area = row['kitchen_area']
    living_area = row['living_area']
    total_area = row['total_area']
    if open_plan == True and math.isnan(living_area):
        row['kitchen_area'] = 0.0
        row['living_area'] = total_area
    if open_plan == True:
        row['kitchen_area'] = 0.0
    else:
        kitchen_area
    return row

data = data.apply(new_open_plan, axis=1)
```

Город Санкт-Петербург широко известен сохранением обънектов недвижимости дореволюционной постройки (до 1917 года), которые, в силу тогдашних архитектурных требований, зачастую имели высоту потолков от 3м до 4,5м.

Таким образом, обобщая полученные выводы, для дальнейше работы с датасетом целесообразно оставить диапазон данных 2,3м - 4,5м, а остальные значения исключить из выборки, кроме значений в диапазоне между 23-35, которые нужно заменить, сместив один знак десятичной дроби влево.

Напишем функцию для строки для замены значений в столбце ceiling_height в диапазоне 18-35м:

```python
def height(row):
    ceiling_height = row['ceiling_height']
    if 18 < ceiling_height < 35:
        return ceiling_height/10.0
    else:
        return ceiling_height

data['ceiling_height'] = data.apply(height, axis=1)

# проверим диапазон 20-35м

data.query('20.0 < ceiling_height < 35.0') 
```

### Проверка столбца `locality_name` на дубликаты и категоризация локаций

Проверим уникальные значения в locality_name:

Есть смысловые дубликаты (поселок и посёлок), заменим их:

```python
data = data.replace(['поселок', 'поселок городского типа'], ['посёлок', 'посёлок городского типа'])
```

Для удобства работы с данными категоризируем локации - разделелим их на группы при помощи функции:

```python
def local(row):
    city = row['locality_name']
    try:
        if 'Санкт-Петербург' in city:
            return city
        if 'село' in city:
            return 'село'
        if 'деревня' in city:
            return 'деревня'
        if 'город' in city:
            return 'пгт'
        if 'посёлок' in city:
            return 'посёлок'
        if 'товарищ' in city:
            return 'садовое товарищество'
        else:
            return 'города ЛО'
    except:
        return 'Не заполнено'
    
data['local_group'] = data.apply(local, axis=1)
```

Изменим порядок столбцов:

Проверим датафрейм:

На мой взгляд, территориальная привязка локации квартиры через парки/водоемы достаточно точный способ восстановления такого рода данных. А значит, если суметь воспользоваться этой информацией и дополнить отсутствующие данные полученными сведениями, исследование будет более взвешенным.

На практике мы знаем, что стоимость жилья между типами населенных пунктов (городом и посёлком и т.п.), между районами одного населенного пункта (районное деление городов) и даже между локациями внутри одного района (например, центр города и весь административный центральный район или Московский проспект и весь Московский район) может отличаться и, как правло, так и бывает. 

В рамках исследования я буду считать, что наличие/отсуствие парков/водоемов влияет на стоимость квартиры, а значит, если сравнить стоимость конкретной квартиры без локации с медианной стоимостью квартиры конкретной группы локаций, то можно получить достаточно точное определение предполагаемой локации такой квартиры.

Построим сводную таблицу по группам локаций и рассчитаем медианную стоимость:

```python
data_local = data.pivot_table(index=['local_group'], values='last_price', 
                              aggfunc=['count', 'median']).astype('int')
              
data_local.columns = ['count', 'median_last_price']

data_local = data_local.reset_index()

data_local
```

Построим сводную таблицу по паркам и водоемам и определим медианную стоимость жилья по группам локаций:

```python
pivot_parks_ponds = (data.pivot_table(index=['local_group'], columns=['parks_around3000', 'ponds_around3000'],
                                     values='last_price', aggfunc='median', fill_value=0)
                     .query('local_group != "Не заполнено"'))

pivot_parks_ponds
```

Создадим функцию со вложенным циклом, которая будет сравнивать стоимость жилья в объявлении с ближайшим 
медианным значением стоимости по сводной таблице для заполнения локации:

```python
def func(row):
    parks = row['parks_around3000']
    ponds = row['ponds_around3000']
    price = row['last_price']
    location = row['local_group']
    
    if location == "Не заполнено":
        local_group_medians = pivot_parks_ponds[parks][ponds]

        min_diff = None
        best_location = None
        for local, median_price in local_group_medians.items():
            price_diff = abs(price - median_price)
            if min_diff == None or min_diff > price_diff:
                min_diff = price_diff
                best_location = local      
        row['local_group'] = best_location
                
    return row
    
data = data.apply(func, axis=1)
    
# проверка заполнения локаций по группам

display(data.query('local_group == "Не заполнено"')) 
```

При наличии водоема поблизости стоимость квартиры на первом или другом этаже вырастает всего на 3%-4% , тогда как стоимость квартиры на последнем этаже увеличиается на 10% (по сравнению с квартирами, рядом с которыми их нет). При последующем увеличении числа водоемов рост стотмости квартиры не идентичен: при наличии двух водоемов прирост стоимости квартиры составит от 6% до 15%, а при наличии трех водоемов прирост стоимости составит от  7% до 29%

### Анализ влияния удаленности от центра

Напишем функцию для категоризации расстояния квартир до центра, значения в которой будем читать значения как "до ... м" то есть, 500 = "до 500 м"

```python
display(good_data['cityCenters_nearest'].describe())
```

```python
def center_nearest(row):
    cityCenters_nearest = row['cityCenters_nearest']
    if cityCenters_nearest <= 1000:
        return 1000
    if 1001 <= cityCenters_nearest <= 2000:
        return 2000
    if 2001 <= cityCenters_nearest <= 3000:
        return 3000
    if 3001 <= cityCenters_nearest <= 5000:
        return 5000
    if 5001 <= cityCenters_nearest <= 10000:
        return 10000
    if 10001 <= cityCenters_nearest <= 15000:
        return 15000
    if 15001 <= cityCenters_nearest <= 20000:
        return 20000
    if 20001 <= cityCenters_nearest <= 25000:
        return 25000
    if 25001 <= cityCenters_nearest <= 30000:
        return 30000
    if 30001 <= cityCenters_nearest <= 35000:
        return 35000
    if 35001 <= cityCenters_nearest <= 40000:
        return 40000
    if 40001 <= cityCenters_nearest <= 45000:
        return 45000
    if 45001 <= cityCenters_nearest <= 50000:
        return 50000
    if 50001 <= cityCenters_nearest <= 55000:
        return 55000
    if 55001 <= cityCenters_nearest <= 60000:
        return 60000
    if cityCenters_nearest >= 60001:
        return 65000
    else:
        return cityCenters_nearest
    
good_data['center_nearest'] = good_data.apply(center_nearest, axis=1)

good_data.head()
```

Построим столбчатую диаграмму:

```python
plt.show(good_data.groupby('center_nearest')['last_price'].mean().plot.bar(grid=True))
```

```python
good_data['last_price'].corr(good_data['center_nearest']).round(2)
```

График показывает обратную зависимость стоимости квартиртиры от удаленности от центра: чем больше увеличивается расстояние от квартиры до центра города, тем ниже становится стоимость картиры.
Обратная зависимость также подтверждается отрицательным коэффициентом Пирсона, однако, значени коэффициента всего -0,33, поэтому зависимость определяется как слабая.

###  Анализ влияния высоты потолков в центре Санкт-Петербурга

Сгруппируем значения высоты потолков через функцию:

```python
def height_group(row):
    ceiling_height = row['ceiling_height']
    if ceiling_height <= 2.5:
        return 2.5
    if 2.5 < ceiling_height <= 2.7:
        return 2.7
    if 2.7 < ceiling_height <= 3.0:
        return 3.0
    if 3.0 < ceiling_height <= 3.3:
        return 3.3

data_stp_center['ceiling_height_group'] = data_stp_center.apply(height_group, axis=1)

data_stp_center.head()
```

Рассчитаем коэффициент корреляции Пирсона:

```python
data_stp_center['ceiling_height'].corr(data_stp_center['last_price']).round(2)
```

На диаграмме видим положительную тенденцию увеличения стоимости при увеличивающемся значении высоты потолков квартиры. Положительный коэффициентр корреляции Пирсона, равный 0,2, говорит о том, что зависимость ниже среднего, скорее стремящаяся к слабой. **В 5-ом разделе выполнялся анализ факторов, влияющих на стоимость квартир.** 

Для поиска взаимосвязи между различными параметрами квартир использовались диаграммы (ячеечные `hexbin`, рассеивания `scatter`, столбчатые `bar`) и рассчитывались коэффициенты корреляции Пирсона `corr()`.

В рамках исследования показатели коэффициента Пирсона было решено рассматривать следующим образом:
1. 0.0 - 0.09 по модулю - связь отсутствует
2. 0.1 - 0.39 по модулю - связь слабая, ниже среднего
3. 0.4 - 0.59 по модулю - связь средяя, слабо-устойчивая
4. 0.6 - 0.89 по модулю - связь крепкая, существенная, выше среднего
5. 0.9 - 1.0 по модулю - связь абсолютная

Также в исследовании используется такая характеристика зависимости как положительная или отрицательная. Если при увеличении одной величины растет и вторая, то взаимосвязь положительная, если же при росте одной из величин вторая уменьшается, то связь отрицательная.

Ниже расположены в порядке убывания факторы, наиболее значимые для определения стоимоcти квартиры, и указана степень их взаимосвязи со стоимостью продажи:
1. Площадь объекта - зависимость положительная, выше среднего (0.8)
2. Стоимость за квадратный метр - зависимость положительная, выше среднего (0.61)
3. Количество комнат - зависимость положительная, средняя (0.5)
4. Удаленность объекта от центра - зависимость отрицательная, ниже среднего (-0.34)

Среди прочего, были рассмотрены такие факторы, как:
1. Этажность квартир (первый, последний, другой). 
2. Наличие в радиусе 3км парков и водоемов.
3. Время публикации объявления: день недели, месяц и год.

Для последней группы факторов расчет коэффициента Пирсона в основном невозможен, поэтому определить степень зависимости несколько затруднительно, однако, анализ совершенно четко определяет её наличие и не линейный, а сложный, даже взаимозависимый характер. 

Так, например, зависимость стоимости квартир только от **_категории этажности и от категории этажности в связке с наличием в радиусе 3000м парков/водоемов_** ведет себя по-разному: в первом случае доля квартир на последнем этаже увеличивается несколько более низкими темпами. Кроме того, при наличии поблизости парков/водоемов  в зависимости от их числа (от 1 до 3)прирост стоимости внутри категорий этажности разный, а в некоторых случаях цена даже снижается на 0,5% (подробнее указано ниже).

При наличии парка поблизости стоимость квартиры резко возрастает на 12%-19% (по сравнению с квартирами, рядом с которыми их нет), при последующем увеличении числа парков поблизости рост в цене не идентичен: при наличии двух парков прирост стоимости квартиры составит от 3% до 8%, а при наличии трех - от  4% до 12%.

При наличии водоема поблизости стоимость квартиры на первом или другом этаже вырастает всего на 3%-4% , тогда как стоимость квартиры на последнем этаже увеличиается на 10% (по сравнению с квартирами, рядом с которыми их нет). При последующем увеличении числа водоемов рост стотмости квартиры не идентичен: при наличии двух водоемов прирост стоимости квартиры составит от 6% до 15%, а при наличии трех водоемов прирост стоимости составит от  7% до 29%

**_В отношении фактора времени публикации_** данные рассматривались в разрезе дня недели, месяца и года.
Несмотря на минимальные изменения стоимости, было решено считать, что зависимость между стоимостью квартиры и днем недели,а также годом, в который была размещена публикация, отсутствует, а зависимость между стоимостью квартиры и  месяцем, в который была размещена публикация, существует, но слабая, скорее связанная с некоторой сезонностью в жизни города.

Пик стоимости приходится на сентябрь, что, вероятнее весго, вызвано началом учебного года и значительным увеличеснием числа иногородних учащихся. После падения повышенного спроса в октябре стоимость несколько проседает до 5,8 млн, но уже с ноября выравнивается под среднее значение 6 млн. Сниже стоимости в июне можно связать с началом сезоном отпусков и дачным периодом, из-за чего спрос падает.
Наиболее привлекательные для продажи квартир периоды - это сентябрь, а наименее - июнь и октябрь.
